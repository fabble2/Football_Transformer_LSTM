```python
import pandas as pd
import pickle
from sklearn import linear_model
import numpy as np
import sklearn
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
''' Quotenausgabe von allen Mannschaften für das naechste Spiel
    spieldaten: pkl Datei
    anfang: Spieltag bei dem man beginnen
    ende: Spieltag bis wann man schauen soll'''

def Spieltag(spieldaten, anfang,ende,anfrage):
    
    anfang= anfang*9
    ende= ende*9
    dfmod= einlesen(spieldaten,anfang,ende)
    spiele= dfmod.values

    anfragespalte= art(spiele,anfrage)
    count=0
    for i in anfrage:
        dfmod[i]= anfragespalte[count]
        count=count+1
    df_model= tabelle(anfrage,anfragespalte,dfmod,anfang,ende)

    
    '''Verwendung von Ridge Regression
        wird verwendet wenn es viele Features besitzt oder wenige Trainingsdaten besitzt
        Vermeidung von Un- oder Ueberanpassung'''
    
    #print(df_model)
    lr = sklearn.linear_model.Ridge(alpha=0.001) 
    X = df_model.drop(anfrage,axis=1)
    #y-Achse bilden
    if len(anfrage) ==1:
        y= df_model[anfrage[0]]
    else:
        y = np.vstack((df_model[anfrage[0]],df_model[anfrage[1]]))
        for i in range(2,len(anfrage)):
            if len(anfrage) <=2:
                break
            else:
                y = np.vstack((y,df_model[anfrage[i]]))
            
    y = y.transpose()
    y = ytransform(y,"PCA")

    y = y.reshape(y.shape[0])
    lr.fit(X, y)

    df_ratings = pd.DataFrame(data={'team': X.columns, 'rating': lr.coef_})

    ''' Ausgabe vorbereiten: Mannschaft1|Mannschaft2|Endergebnis|Quote1|Quote2 
        Das naechste Spiel wird angezeigt mit dessen Quoten und dem Ergebnis'''

    coef= df_ratings.values
    #print(coef)
    #Nächstes Spiel
    #Tabelle vorbereiten
    anfang=ende
    ende=ende+9
    next_game,ergebnisse= einlesen_spiele(spieldaten,anfang,ende)
    next_tabelle= tabellespiele(next_game)
    print(X.shape)
    print(next_tabelle.shape)
    prediction = lr.predict(next_tabelle)
    ergebnis_liste=ergebnisse.values.tolist()

    gesamt= ausgabe(next_game,ergebnis_liste,prediction)

    for i in gesamt:
        i[0]= i[0].decode()
        i[1]= i[1].decode()
        #print(i)
    return gesamt
liste= Spieltag("Spiele2018.pkl",0,1,["siege","differenz"])
for i in liste:
    print(i)
```

```python
def einlesen(spieldaten, anfang, ende):
    df = pd.read_pickle(spieldaten)

    #Tabelle besteht aus Mannschaft1,2 und dem Endergebnis
    dfmod= df[["Mannschaft1", "Mannschaft2", "Endergebnis"]][anfang:ende]
    return dfmod
def einlesen_spiele(spieldaten,anfang,ende):
    df = pd.read_pickle(spieldaten)
    dfmod= df[["Mannschaft1", "Mannschaft2"]][anfang:ende]
    dfmod_ergebnisse= df[["Endergebnis"]][anfang:ende]
    return dfmod, dfmod_ergebnisse
    
```

```python
def art(spiele,anfrage):
    '''Tordifferenz modifizieren zu Siege/Niederlage'''
    gesamt=[]
    for i in anfrage:
        
        if str(i)=="siege":
            siege=[] # groesser 0= Heim gewonnen, kleiner 0= Gast gewonnen
            for i in spiele:
                ergebnis= i[2]
                ergebnis= ergebnis.split(":")
                sieg=int(ergebnis[0])-int(ergebnis[1])
                if sieg >0:
                    sieg=1
                elif sieg <0:
                    sieg=-1
                siege.append(sieg)
            #print(differenz)
            gesamt.append(siege)
        elif str(i)=="differenz":

            differenz=[] # groesser 0= Heim gewonnen, kleiner 0= Gast gewonnen
            for i in spiele:
                ergebnis= i[2]
                ergebnis= ergebnis.split(":")
                diff=int(ergebnis[0])-int(ergebnis[1])
                differenz.append(diff)
            gesamt.append(differenz)
    return gesamt
```

```python
def tabelle(anfrage, anfragespalte, dfmod,anfang,ende):
    if (ende-anfang)==9:
        df_model= tabellespiele(dfmod)
        df_model[anfrage] = dfmod[anfrage]
        return df_model
    dfmod[anfrage]=anfragespalte

    '''Heimvorteil beruecksichtigen'''
    dfmod['heimsieg'] = np.where(dfmod[anfrage] > 0, 1, 0)
    dfmod['heimniederlage'] = np.where(dfmod[anfrage] <0,1,0)

    '''Dummy Variabeln erstellen
       Stellen Gast- und Heimmannschaft vor
       df_visitor: Zeigt bei jedem Spiel die Gastmannschaft
       df_home:Zeigt die Heimmannschaft von jedem Spiel'''

    df_visitor = pd.get_dummies(dfmod['Mannschaft2'], dtype=np.int64)
    df_home = pd.get_dummies(dfmod['Mannschaft1'], dtype=np.int64)


    '''Fehlerkorrektur in der Tabelle (NaN Fehler)'''
    nan_home=[]
    for (columnName, columnData) in df_home.iteritems():
       team= columnName
       if np.sum(columnData.values)>1:
            teamwerte= [team,columnData.values]
            nan_home.append(teamwerte)

    nan_gast=[]
    for (columnName, columnData) in df_visitor.iteritems():
       team= columnName
       if np.sum(columnData.values)>1:
            columnData = columnData*-1
            teamwerte= [team,columnData.values]
            nan_gast.append(teamwerte)
            #nan_gast.append(columnData.values)        

    '''Transformation: 1:Heimmannschaft, -1: Gastmannschaft und hinzunahme von Differenz'''
    df_model = df_home.sub(df_visitor)
    df_model[anfrage] = dfmod[anfrage]

    '''
    Wenn eine Mannschaft nur Heimspiele hat muessen die Werte umgeschrieben,
    sonst nan in der Spalte und man kann nicht weiter arbiten
    '''
    count=0
    for (columnName, columnData) in df_model.iteritems():

        if np.isnan(np.sum(columnData.values)):
            for i in nan_home:
                if columnName==i[0]:
                    df_model[columnName]= i[1]
            for i in nan_gast:
                if columnName==i[0]:
                    df_model[columnName]=i[1]
    return df_model
```

```python
def tabellespiele(dfmod):
    df_visitor = pd.get_dummies(dfmod['Mannschaft2'], dtype=np.int64)
    df_visitor= df_visitor*-1
    df_home = pd.get_dummies(dfmod['Mannschaft1'], dtype=np.int64)

    df_model=pd.concat([df_home,df_visitor],axis=1)
    return df_model
```

```python
def ausgabe(spiele, ergebnis_liste,prediction):
    mannschaft1=[]
    mannschaft2=[]
    ergebnisse=[]
    gesamt=[]
    for i in spiele.iterrows():
        mannschaft1.append(i[1][0])
        mannschaft2.append(i[1][1])
    for i in ergebnis_liste:
        ergebnisse.append(i[0])
    
    for i in range(0,9):
        game=[mannschaft1[i],mannschaft2[i],ergebnisse[i],prediction[i]]
        gesamt.append(game)
    
    return gesamt
    
    
```

```python
def ytransform(y,art):
    if art=="TSNE":
        return TSNE(n_components=1).fit_transform(y)
    elif art=="PCA":
        return PCA(n_components=1).fit_transform(y)
    return "Fehler"
```
