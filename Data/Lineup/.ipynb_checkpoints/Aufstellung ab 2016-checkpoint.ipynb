```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import pickle
import re
```

```python
#Jede Seite hat zu jedem Spiel einen Link. 
#Diesen holen und dann pro Spiel die Aufstellung holen 
def spieltage(url):
    counter=1
    req = requests.get(url)
    soup = BeautifulSoup(req.text,'html.parser')
    liste=[]
    
    if len(url)==47:
        partlink= url[28:47] #bundesliga/2019/1/ oder andere Zahl am Ende
    elif len(url)==48:
        partlink= url[28:48] #Zahl groesser 9
    
    for link in soup.find_all('a'):
        possible_spieltag= link.get("href")
        
        if possible_spieltag.startswith(partlink) and "-" in possible_spieltag:
            
            newlink= "https://www.fussballdaten.de/"+possible_spieltag
            liste.append(newlink)
            
    return liste
    
```

```python
#Spiel(Heim und Gast Mannschaften) für einen Spieltag sammeln
def aufstellung(spieltag):
    team1=[]
    team2=[]
    team1ersatz=0
    team2ersatz=0
    #print(spieltag)
    spieltagliste=[]
    
    for i in spieltag:
        team1=[]
        team2=[]
        ersatzspielerteam1=[]
        ersatzspielerteam2=[]
        
        #mannschaften aus link bekommen /mannschaft1-mannschaft2/
        mannschaften= i.split("/")
        while("" in mannschaften):
            mannschaften.remove("")
        spieltag=mannschaften[len(mannschaften)-2]
        #print(spieltag)
        mannschaften=mannschaften[len(mannschaften)-1]
        mannschaften= mannschaften.split("-")
        
        #aufstellungen, startelf(ersten 22 Spieler)
        #ersatzelf(jeder zweite Spieler danach, Ueberpruefung in welcher Mannschaft)
        url=i+"aufstellung/"
        #print(url)
        req = requests.get(url)
        soup= BeautifulSoup(req.text, "html.parser")
        #print(soup.prettify())
        spielername = soup.findAll('a', attrs={'class':"text lineup"})
        
        #print(len(spielername))
        counter=0
        for i in spielername:
            splitter= str(i).split(">")
            while("" in splitter):
                splitter.remove("")
            name= splitter[len(splitter)-1][:-3]
            #print(name)
            if counter<11:
                team1.append(name)
                counter= counter+1
                continue
            elif counter<22:
                team2.append(name)
                counter= counter+1
                continue
        
        #ersatzspieler
        ersatzspielerheim= soup.find("div", attrs={'class': "col-md-6 col-sm-6 col-xs-12 heim-content"})
        newersatzspielerheim= str(ersatzspielerheim)
        spielerlisteheim= newersatzspielerheim.split("<span")
        counter2= 0
        for i in spielerlisteheim:
            #print(i)

            if 'fa fa-exchange' in i:
                #print(spielerliste[counter2-1])
                ename= spielerlisteheim[counter2-1].split(">")
            
                while("" in ename):
                    ename.remove("")
                ersatzspielerteam1.append(ename[len(ename)-1][:-3])
                
            counter2 =counter2+1
            
        ersatzspielergast= soup.find("div", attrs={'class': "col-md-6 col-sm-6 col-xs-12 gast-content"})
        newersatzspielergast= str(ersatzspielergast)
        spielerlistegast= newersatzspielergast.split("<span")
        counter3= 0
        nextline= 0
        for i in spielerlistegast:
            
            #if mannschaften[0]=="vfbstuttgart":
                #print("")
                #print(i)
                
            if nextline==1:
                ename= i.split(">")
                #print(ename)
                while("" in ename):
                    ename.remove("")
                ersatzspielerteam2.append(ename[len(ename)-1][:-3])
                nextline=0
        
            if 'fa fa-exchange' in i:
                
                ename= i.split(">")
                while("" in ename):
                    ename.remove("")
                
                if "<" in ename[len(ename)-1][:-3]:
                    nextline=1
                else:
                    ersatzspielerteam2.append(ename[len(ename)-1][:-3])
            counter3 =counter3+1
        
        #print(ersatzspielerteam1)
        #print(ersatzspielerteam2)
        team1= team1+ersatzspielerteam1
        team2= team2+ersatzspielerteam2
        
        tempspiel= [spieltag, mannschaften[0], mannschaften[1], team1, team2]
        #print(tempspiel)
        spieltagliste.append(tempspiel)
        
    return spieltagliste
        
#print(spieltag1)
#aufstellung(spieltage("https://www.fussballdaten.de/bundesliga/2016/23/"))
```

```python
#Fuer jede Saison alle Spiele mit Aufstellungen holen#
saison=[]
zeile= ["Spieltag","Mannschaft1","Mannschaft2","Team1","Team2"]

def alleSpieleProSaison():
    saison=[]
    
    firstgame=0
    daten = pd.DataFrame(columns = zeile)
    daten2 = pd.DataFrame(columns = zeile)
    
    for i in range(1,2):
        link= "https://www.fussballdaten.de/bundesliga/2016/"+str(i)+"/"
        print(link)
        spiele= spieltage(link)
        #print(spiele)
        if i<23:
            games=aufstellung(spiele)
        else:
            games=aufstellung2016neu(spiele)
        saison.append(games)
        
        firstgame= firstgame+0
        
        for spiele in saison:
            for spiel in spiele:
                
                if firstgame==0:
                    daten= pd.DataFrame([spiel],columns=zeile)
                    firstgame=1
                else:           
                    daten2= pd.DataFrame([spiel], columns=zeile)
                daten= pd.concat([daten,daten2])
        
        firstgame=1
        saison=[]
    daten= daten.reset_index()
    daten.to_pickle("./Aufstellungen2016.pkl")
    return daten.reset_index()

#alleSpieleProSaison()
```
