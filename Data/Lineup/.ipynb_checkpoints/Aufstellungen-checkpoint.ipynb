```python
#Jede Seite hat zu jedem Spiel einen Link. Diesen holen und dann pro Spiel die Aufstellung holen 
def spieltage(url):
    counter=1
    req = requests.get(url)
    soup = BeautifulSoup(req.text,'html.parser')
    liste=[]
    
    if len(url)==47:
        partlink= url[28:47] #bundesliga/2019/1/ oder andere Zahl am Ende
    elif len(url)==48:
        partlink= url[28:48] #Zahl groesser 9
    
    for link in soup.find_all('a'):
        possible_spieltag= link.get("href")
        
        if possible_spieltag.startswith(partlink) and "-" in possible_spieltag:
            
            newlink= "https://www.fussballdaten.de/"+possible_spieltag
            liste.append(newlink)
            
    return liste

#spieltag1= spieltage("https://www.fussballdaten.de/bundesliga/2019/10/")    
```

```python
#Alle Spieler die beim aktuellen Spieltag gespielt haben, separtiert in Partien und diesen 2 Mannschaften
def aufstellung(spieltag):
    
    #print(spieltag)
    spieltagliste=[]
    for i in spieltag:
        
        #mannschaften aus link bekommen /mannschaft1-mannschaft2/
        mannschaften= i.split("/")
        while("" in mannschaften):
            mannschaften.remove("")
        spieltag=mannschaften[len(mannschaften)-2]
        #print(spieltag)
        mannschaften=mannschaften[len(mannschaften)-1]
        mannschaften= mannschaften.split("-")
        #print(mannschaften)
        
    
        #aufstellungen, startelf
        url=i+"aufstellung/"
        #print(url)
        req = requests.get(url)
        soup= BeautifulSoup(req.text, "html.parser")
        #print(soup.prettify())
        spielername = soup.findAll('span', attrs={'class':'name'})
        team1= []
        team2= []
        counter=0
        for i in spielername:
            if counter <11:
                team1.append(i["title"])
                counter= counter+1
            else:
                team2.append(i["title"])
                counter= counter+1
        
        #ersatzspielerheim
        ersatzspielerteam1=[]
        ersatzspieler= soup.find("div", attrs={'class':"col-md-6 col-sm-6 col-xs-12 heim-content"})
        newersatzspieler= str(ersatzspieler)
        spielerliste= newersatzspieler.split("<span")
        counter2= 0
        for i in spielerliste:
            #print(i)

            if 'fa fa-exchange' in i:
                #print(spielerliste[counter2-1])
                ename= spielerliste[counter2-1].split(">")
                while("" in ename):
                    ename.remove("")
                ersatzspielerteam1.append(ename[len(ename)-1][:-3])
            counter2 =counter2+1

        #ersatzspielergast
        ersatzspielerteam2=[]
        ersatzspieler_gast= soup.find("div", attrs={'class':"col-md-6 col-sm-6 col-xs-12 gast-content"})
        new_ersatzspieler_gast= str(ersatzspieler_gast)
        spielerliste_gast= new_ersatzspieler_gast.split("<span")
        counter2= 0
        for i in spielerliste_gast:
            #print("::"+i)

            if 'fa fa-exchange' in i and i.endswith("</a>"):
                #print(spielerliste_gast[counter2+1])
                ename_gast= spielerliste_gast[counter2].split(">")
                while("" in ename_gast):
                    ename_gast.remove("")
                ersatzspielerteam2.append(ename_gast[len(ename)-1][:-3])
            elif 'fa fa-exchange' in i and len(spielerliste)>=counter2+1 and spielerliste_gast[counter2+1].endswith("</a>"):
                #print(spielerliste_gast[counter2+1])
                ename_gast= spielerliste_gast[counter2+1].split(">")
                while("" in ename_gast):
                    ename_gast.remove("")
                ersatzspielerteam2.append(ename_gast[len(ename)-1][:-3])
            counter2 =counter2+1
        
        team1= team1+ersatzspielerteam1
        team2= team2+ersatzspielerteam2
        
        print(team1)
        print(team2)
        
        tempspiel= [spieltag, mannschaften[0], mannschaften[1], team1, team2]
        #print(tempspiel)
        spieltagliste.append(tempspiel)
        
        
        
    return spieltagliste
        
#print(spieltag1)
#aufstellung(spieltage("https://www.fussballdaten.de/bundesliga/2016/1/"))
        
```

```python


#Fuer jede Saison alle Spiele mit Aufstellungen holen#
saison=[]
zeile= ["Spieltag","Mannschaft1","Mannschaft2","Team1","Team2"]

def alleSpieleProSaison():
    saison=[]
    
    firstgame=0
    daten = pd.DataFrame(columns = zeile)
    daten2 = pd.DataFrame(columns = zeile)
    
    for i in range(21,22):
        link= "https://www.fussballdaten.de/bundesliga/2016/"+str(i)+"/"
        #print(link)
        spiele= spieltage(link)
        #print(spiele)
        games=aufstellung(spiele)
        saison.append(games)
        
        firstgame= firstgame+0
        for spiele in saison:
            for spiel in spiele:
                
                if firstgame==0:
                    daten= pd.DataFrame([spiel],columns=zeile)
                    firstgame=1
                else:           
                    daten2= pd.DataFrame([spiel], columns=zeile)
                daten= pd.concat([daten,daten2])
        
        firstgame=1
        saison=[]
    daten= daten.reset_index()
    daten.to_pickle("./Aufstellungen2016.pkl")
    return daten.reset_index()

#alleSpieleProSaison()
```
