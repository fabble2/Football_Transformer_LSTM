```python
import pandas as pd
from collections import defaultdict
from difflib import SequenceMatcher
import difflib
from unidecode import  unidecode
from urllib.parse import unquote
import pickle

#Die zwei pkl Dateien einlesen
#daten: Enthält die Aufstellungen vom Spieltag
#fifa: Enthält die GES Werte von den Spielern der Mannschaften am Spieltag
def einlesen(jahr):
    daten=pd.read_pickle("Aufstellungen20"+str(jahr)+".pkl")
    fifa=pd.read_pickle("Fifa"+str(jahr)+".pkl")
    fifa=fifa.iloc[::-1]
    return daten,fifa

#dict erstellen welches pro Aufstellung am jeweiligen Spieltag 
#die Aufstellung+Mannschaftsname erstellt
def spieltag_daten_erstellen(daten):
    #Nummer 33= 1.Spieltag,..., Nummer 0= 34.Spieltag
    spieler=[]
    spieltag_daten={}
    teamnames_daten=[]
    spieltagsnummer=33
    for row in daten.iterrows():
        rowlist=row[1].tolist()

        nummer=34-int(rowlist[1])
        if spieltagsnummer==nummer:
            liste1= rowlist[4]
            liste1.append(rowlist[2])

            liste2= rowlist[5]
            liste2.append(rowlist[3])

            spieler.append(liste1)
            spieler.append(liste2)

        else:
            spieltag_daten.update({spieltagsnummer:spieler})

            spieler=[]
            teamnames_daten=[]
            liste1= rowlist[4]
            liste1.append(rowlist[2])

            liste2= rowlist[5]
            liste2.append(rowlist[3])

            spieler.append(liste1)
            spieler.append(liste2)

            spieltagsnummer= spieltagsnummer-1
    spieltag_daten.update({spieltagsnummer:spieler})
    
    return spieltag_daten
#dict erstellen welches GES Werte+Spielername+ Mannschaft einspeichert
def spieltag_fifa_erstellen(fifa):
    liga=[]
    teamnames_fifa=[]
    for i in fifa.to_dict().items():
        team=i[0]
        liste= i[1]
        for j in range(33,-1,-1):
            listehelp= liste[j]
            listehelp.append(team)
            liga.append({j:listehelp})


    spieltag_fifa={}

    for i in range(33,-1,-1):
        zahl=i%33
        mannschaften=[]
        for mannschaft in liga:
            for spieltage,spieler in mannschaft.items():
                if spieltage==i:
                    #print(spieler)
                    mannschaften.append(spieler)

        spieltag_fifa.update({i:mannschaften})
    return spieltag_fifa

#Ähnlichkeitssuche, welche Mannschaft wird am Spieltag betrachtet    
def similar(a, b): 
    return SequenceMatcher(None, a, b).ratio()

#Mannschaft finden, welche betrachtet wird
#Spielerwerte und Teamname abspeichern
#für jedes Spiel pro Spieltag und für jeden Spieltag pro Saision
def result_erstellen(spieltag_daten,spieltag_fifa):
    mannschaften=[] 
    for mannschaft in spieltag_fifa[1]: 
        team= mannschaft[len(mannschaft)-1] 
        team= unidecode(unquote(team)) 
        mannschaften.append(team)

    result={} 
    team_ges=[] 
    count=1 
    for key,value in spieltag_daten.items():

        for spieler in value:
            aufstellung= spieler
            mannschaft= spieler[len(spieler)-1]
            wert=0
            spieltag={}
            mannschaft_ges=[]
            for liste in spieltag_fifa[key]:
                mannschaft2= liste[len(liste)-1]
                mannschaft2=unidecode(unquote(mannschaft2))
                similarity= similar(mannschaft,mannschaft2)

                if wert < similarity:
                    wert= similarity
                    #print(liste)
                    for i in liste:
                        name= str(i)[:-3]
                        ges= str(i)[-2:]
                        if name in aufstellung:
                            mannschaft_ges.append(ges)

            spieltag.update({mannschaft:mannschaft_ges[:11]})
            team_ges.append(spieltag)
        result.update({count:team_ges})
        count +=1
    return result

#dict in pkl speichern
def pkl_erstellen(result,jahr):
    data= pd.DataFrame.from_dict(result)
    filename= "GES"+str(jahr)+".pkl"
    file = open(filename, 'wb')
    pickle.dump(data,file)
    file.close()

    
jahr=19
daten,fifa= einlesen(jahr)
spieltag_daten= spieltag_daten_erstellen(daten)
spieltag_fifa= spieltag_fifa_erstellen(fifa)
result= result_erstellen(spieltag_daten,spieltag_fifa)
pkl_erstellen(result,jahr)
```
