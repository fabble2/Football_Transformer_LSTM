```python
import numpy as np
import pandas as pd
from collections import defaultdict
from difflib import SequenceMatcher
import difflib
from unidecode import  unidecode
from urllib.parse import unquote
import pickle
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import minmax_scale
```

```python
#Eine Mannschaft hat pro Spiel einen eigenen Vektor
#besteht aus Tabellenwerte, Aufstellung und Werte des Teams (Teil 1)
#desweiteren Infos zur Vergangenheit(Spielstatistiken) (Teil 2)

#Ein Spiel besteht aus 2 Vektoren: Mannschaft A und Mannschaft B
#Zusätzlich kommt noch ein Wettquotenvektor hinzu: Werte wie Spiel ausgehen wird

#Die 18 als dict oder enum speichern, quasi key= mannschaftsname, value= matrixnummer
matrix= np.zeros((18,34,67))
 
```

```python
#input: Mannschaft und Spiteltag
#Tabellenwerte in dict, key= Spieltag(1-34) und dann Mannschaft suchen
#Wird mit Hilfsmethode Similar gefunden 
#output: Tabbelnplatz, Siege, Niederlade, Unentschieden, Differenz als Liste zurückgegeben

def Tabellenwerte(nummer,mannschaft):
    
    if nummer == 0:
        return [0,0,0,0,0]
    else:
        tabelle_rohdaten= pd.read_pickle("Tabelle19.pkl")
        tabelle= tabelle_rohdaten.to_dict()
        spieltag= tabelle[nummer]
        teams= spieltag.keys()
        current_mannschaft=similar(mannschaft,teams)
        return spieltag[current_mannschaft]

#input:Mannschaft(die gefunden werden soll), Liste(alle Mannschaften)
#Ähnlichkeitssuche wird durchgeführt
#output: größte Ähnlichkeit ==> Mannschaftsname in Form der Liste
def similar(mannschaft, liste):
    wert=0
    currentteam="nichts"
    for team in liste:
        similarity= SequenceMatcher(None, mannschaft, team).ratio()
        #print(team, similarity)
        if wert < similarity:
            wert= similarity
            currentteam= team
    #print(currentteam)        
    return currentteam

#Tabellenwerte(2,"bayern")
```

```python
#parameter: Mannschaft und Spieltag
#Werte in dict, key= Mannschaft und dann Spieltag suchen(0-33)
#==> nummer= nummer-1 damit 0 klappt

'''def Umlaute(liste):
    liste= list(liste)
    for i in range(len(liste)):
        name= liste[i]
        liste[i]= unidecode(unquote(name))

    return liste'''
    
def Geswerte(nummer,mannschaft):
    ges_rohdaten= pd.read_pickle("FifaGes19.pkl")
    ges= ges_rohdaten.to_dict()
    teams= (ges.keys())
    current_mannschaft= similar(mannschaft,teams)
    for key,values in ges[current_mannschaft].items():
        if key==(nummer-1):
            return values
    
#Geswerte(10,"leverkusen")

```

```python
#parameter: Mannschaft(1-34) und Spieltag
#Werte in einem dict[1], dort befinden sich alle Spiele(612(34*9*2))
#Diese in 18er Blöcke separieren
#Leider alle Zahlen der mannschaften durchgehen und dann key= mannschaft

def Korrektur(liste):
    for i in liste:
        if str(i)=="leipzig":
            index= liste.index(i)
            liste[index]='70'
    return liste
def Aufstellung(nummer,mannschaft):
    aufstellung_rohdaten= pd.read_pickle("GES19.pkl")

    aufstellung=(aufstellung_rohdaten.to_dict()[nummer])
    
    #print(aufstellung)
    teams= aufstellung.keys()

    current_mannschaft= similar(mannschaft,teams)
    for key,values in aufstellung.items():
        if key==current_mannschaft:
            values= Korrektur(values)
            return values


for i in range(1,35):
    print(Aufstellung(i,"augsburg"))

```

```python
#parameter: speiltag, mannschaft
#jede Zeile enthält spieltagsnummer und zwei mannschaften
#spieltag abfragen und dann reihenweise Mannschaft suchen

def Statistik(nummer,mannschaft):
    
    templiste=[]
    with (open("Statistik2019.pkl", "rb")) as openfile:
        while True:
            try:
                statistik= pickle.load(openfile)
                if statistik[0]==nummer:
                    templiste.append(statistik)
                
            except EOFError:
                break
    teams=[]
    for i in templiste:
        teams.append(i[1])
        teams.append(i[2])
    
    current_mannschaft= similar(mannschaft,teams)
    #print(current_mannschaft)
    for i in templiste:
        if str(i[1])==current_mannschaft:
            liste =i[4]
            for j in liste:
                wert=str(j).replace("km","").strip()
                wert=str(wert).replace("%","").strip()
                wert=str(wert).replace(",",".").strip()
                liste[liste.index(j)]=wert
            return liste, i[2],"heim"
        elif str(i[2])==current_mannschaft:
            liste=i[5]
            for j in liste:
                wert=str(j).replace("km","").strip()
                wert=str(wert).replace("%","").strip()
                wert=str(wert).replace(",",".").strip()
                liste[liste.index(j)]=wert
            return liste, i[1],"gast"
    
    
#Statistik(10,"leverkusen")
```
