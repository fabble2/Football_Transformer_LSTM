```python
"""
input: url
get all odds from the url with help from a webdriver
ouput: list of data (game, odds)
"""
def wettquote(url):

    driver = webdriver.Firefox(executable_path=r'/home/fabian/Schreibtisch/UNI/Projekt/1. Bundesliga/Wettquoten/geckodriver')    
    driver.get(url)
    soup = BeautifulSoup(driver.page_source)
    #print(soup.prettify())
    
    #Wahrscheinlichkeiten
    #jedes Spiel hat 3 Wahrscheinlichkeiten
    table=soup.findAll("td",{'class':['result-ok odds-nowrp','odds-nowrp']})
    game=[]
    odds=[]
    counter=0
    anzahl=0
    relegation=0
    for i in table:
        if relegation>5:
            
            if counter==3 and len(game)==3:
                counter=0
                odds.append(game)
                game=[]
            splitter=str(i).split(">")
            while("" in splitter):
                splitter.remove("")
            game.append(splitter[len(splitter)-2][:-3])
            counter=counter+1
            anzahl=anzahl+1
        relegation=relegation+1
    odds.append(game)
    #print(odds)
    
    #Datum und Spiele holen
    #Datum nicht nÃ¶tig, man zaehlt bis 9 und hat einen Spieltag
    dates= soup.findAll(["tr","td"],{'class':['center nob-border','name table-participant']})
    
    counter3=0
    spiele=[]
    for i in dates:
        
        if counter3>3:
            '''if "center nob-border" in str(i):
                splitter= str(i).split(">")
                while("" in splitter):
                    splitter.remove("")
                for j in splitter:
                    if "2019" in j or "2018" in j:
                        print("")
                        print(j[:-6])'''
            spiel=""
            if "name table-participant" in str(i):
                splitter= str(i).split(">")
                while("" in splitter):
                    splitter.remove("")
                for j in splitter:
                    if "span" in j or "</a" in j:
                        if j!='<span class="bold"' and j!="</a":
                            j= j.replace("</span","")
                            j= j.replace('<span class="bold"',"")
                            j= j.replace("</a","")
                            spiel= spiel+j
                spiele.append(spiel)
            
        counter3 =counter3+1

    driver.quit()
    data=[spiele, odds]
    return data
    
#wettquote("https://www.oddsportal.com/soccer/germany/bundesliga-2018-2019/results/")
```

```python
"""
input: url
get the odds from the other pages
output: Data
"""
def wettquote2(url):

    driver = webdriver.Firefox(executable_path=r'/home/fabian/Schreibtisch/UNI/Projekt/1. Bundesliga/Wettquoten/geckodriver')    
    driver.get(url)
    soup = BeautifulSoup(driver.page_source)
    #print(soup.prettify())
    
    #Wahrscheinlichkeiten
    #jedes Spiel hat 3 Wahrscheinlichkeiten
    table=soup.findAll("td",{'class':['result-ok odds-nowrp','odds-nowrp']})
    game=[]
    odds=[]
    anzahl=0
    counter=0
    for i in table:
        if counter==3:
            counter=0
            odds.append(game)
            game=[]
        splitter=str(i).split(">")
        while("" in splitter):
            splitter.remove("")
        game.append(splitter[len(splitter)-2][:-3])
        counter=counter+1
        anzahl=anzahl+1
    odds.append(game)
    
    #Datum und Spiele holen
    #Datum nicht nÃ¶tig, man zaehlt bis 9 und hat einen Spieltag
    dates= soup.findAll(["tr","td"],{'class':['center nob-border','name table-participant']})
    spiele=[]
    for i in dates:

        spiel=""
        if "name table-participant" in str(i):
            splitter= str(i).split(">")
            while("" in splitter):
                splitter.remove("")
            for j in splitter:
                if "span" in j or "</a" in j:
                    if j!='<span class="bold"' and j!="</a":
                        j= j.replace("</span","")
                        j= j.replace('<span class="bold"',"")
                        j= j.replace("</a","")
                        spiel= spiel+j
            spiele.append(spiel)
    
    #print(spiele)
    driver.quit()
    data=[spiele, odds]
    return data

#wettquote2("https://www.oddsportal.com/soccer/germany/bundesliga-2018-2019/results/#/pages/6")
```
