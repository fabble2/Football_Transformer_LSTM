```python
import numpy as np
import pandas as pd
from collections import defaultdict
from difflib import SequenceMatcher
import difflib
from unidecode import  unidecode
from urllib.parse import unquote
import pickle
import csv
```

```python
#input: pkl Datei 
#in pkl datei nach Spieltagsnummer, Mannschaften suchen
#nehme Mannschaften/Nummer mit Quoten in Form ['-</s', '-</s', '-</s']
#In CSV nach Spiel suchen und Quoten nachreichen


#Alle Fehlerhaften Quoten aus der pkl sammeln
def Fehler_einlesen(jahr):
    templiste=[]
    dokument= "Wettquoten"+str(jahr)+".pkl"
    with (open(dokument, "rb")) as openfile:
        while True:
            try:
                spiel=[]
                wetten= pickle.load(openfile)
                #print(wetten[3])
                if wetten[3]==['-</s', '-</s', '-</s']:
                    spiel.append(wetten[0])
                    spiel.append(wetten[1])
                    spiel.append(wetten[2])
                    templiste.append(spiel)
            except EOFError:
                break
    return templiste
    

#Spieltage holen, in der die fehlerhafte Quote sich befindet
#input: Spielnummer und Saisonjahr
#output: Liste aus [mannschaft1, mannschaft2, [odds]], 9 Einträge (ein Spieltag)

def Korrektur(spieltagsnummer,jahr):
   
    #print(data)
    spieldaten=[]
    ersteZeile=1  
    spieltag= spieltagsnummer
    anfang= (spieltag-1)*9
    ende= spieltag*9
    counter=0
    zahl= str(jahr)
    file= zahl[-2:]+".csv"
    with open(file, 'r') as read_obj:
        csv_reader = csv.reader(read_obj)
        for row in csv_reader:
            
            if ersteZeile==1:
                ersteZeile=0
                continue
            odds=[]
            spiel=[]
            if counter< ende and counter >=anfang:

                odds.append(row[22])
                odds.append(row[23])
                odds.append(row[24])
                #print(row[22],row[23],row[24])
                #print(row[2],row[3],odds)
                spiel.append(row[2])
                spiel.append(row[3])
                spiel.append(odds)
                spieldaten.append(spiel)
            counter+=1
    return spieldaten

#Ähnlichkeitssuche
def similar(mannschaft, liste):
    wert=0
    currentteam="nichts"
    for team in liste:
        similarity= SequenceMatcher(None, mannschaft, team).ratio()
        #print(team, similarity)
        if wert < similarity:
            wert= similarity
            currentteam= team
    #print(currentteam)        
    return currentteam

#die richigen Werte für die fehlerhaften Spiel holen,
#output: Liste mit den berichtigten Spielen
def Ausfuehren(pkl):
    pkl_neu=[]
    for data in pkl:
        csv= Korrektur(data[0],jahr)
        teams=[]
        for mannschaft in csv:
            teams.append(mannschaft[0])
            teams.append(mannschaft[1])
        current=(similar(data[1],teams))
        
        for mannschaft in csv:
            if current in mannschaft:
                data.append(mannschaft[2])
    return pkl
        
#Neues pickle erstellen
#Altes Pickle in List speichern
#Neues Pickle aus alter List, aber die fehlerhaften Spiele durch die berichtigten
#Spiele ersetzt

def create_pickle(jahr,neue_werte):
    sammlung= []
    zahl= str(jahr)
    filename= "Wettquoten"+zahl+".pkl"
    with (open(filename, "rb")) as openfile:
            while True:
                try:
                    wetten= pickle.load(openfile)
                    sammlung.append(wetten)

                except EOFError:
                    break
    for i in sammlung: 
        for j in neue_werte:
            if i[0]== j[0] and i[1]==j[1] and i[2]==j[2]:
                index= sammlung.index(i)
                sammlung[index]=j
    test= sammlung[:307]
    df= pd.DataFrame(test)
    filenamenew= "Wettquoten"+zahl+"Neu.pkl"
    df.to_pickle(filenamenew)
    
    
jahr=2013        
pkl_data = Fehler_einlesen(jahr)    
neue_werte= Ausfuehren(pkl_data)
create_pickle(jahr,neue_werte)
```
