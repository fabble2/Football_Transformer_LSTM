```python
import pandas as pd
import pickle
import torch
import matplotlib.pyplot as plt
import numpy as np


def einlesen_mannschaften(spieldaten):
    count=0
    mannschaften=[]
    with (open(spieldaten, "rb")) as openfile:
        while True:
            try:
                spieltag= pickle.load(openfile)

                if spieltag[1]=="Mannschaft1":
                    continue
                elif count <9:
                    mannschaften.append(spieltag[1])
                    mannschaften.append(spieltag[2])
                    count=count+1
                else:
                    break
            except EOFError:
                break
    return mannschaften

        
def main(daten,anfang,ende,feature):
    dfmod=sammeln(daten,anfang,ende,feature)
    training_input,training_output,test_input,test_output= test_training(dfmod)
    #print(dfmod)
    # in Tensoren umwandeln    
    training_input = torch.FloatTensor(training_input.values) 
    training_output = torch.FloatTensor(training_output.values) 
    test_input = torch.FloatTensor(test_input.values) 
    test_output_tensor = torch.FloatTensor(test_output.iloc[:,1:].values)

    #Eingabedimension= Anzahl der Features
    
    input_size= training_input.size()[1]
    hidden_size= input_size+2 #Anzahl der Knoten
    #print(input_size,hidden_size)
    model= Net(input_size,hidden_size)#Modell erstellen
    criterion = torch.nn.BCELoss() #Lossfunktion: BCE da wir bin채re Ergebnisse haben
    
    #optimizer: enthaelt aktuellen Status und aktualisiert die Parameter basierend 
    #auf den berechneten Verl채ufen
    #ohne momentum paramter
    optimizer = torch.optim.SGD(model.parameters(), lr = 0.9) 
    #mit momentum parameter
    #optimizer = torch.optim.SGD(model.parameters(), lr = 0.9, momentum=0.2)
    
    #Trainingsanfang
    
    model.eval()#Testmodus, keine Gewichte werden gelernt, nur Verlust angezeigt
    y_pred = model(test_input)
    before_train = criterion(y_pred.squeeze(), test_output_tensor)
    #print('Test loss before training' , before_train.item())
    
    model.train()
    epochen = 5000
    errors = []
    for epoch in range(epochen):
        optimizer.zero_grad()
        # Forward pass
        y_pred = model(training_input)
        # Verlustberechnung
        loss = criterion(y_pred.squeeze(), training_output)
        errors.append(loss.item())
        #print('Epoch {}: train loss: {}'.format(epoch, loss.item()))
        
        # Backward berechnet die Gradienten
        loss.backward()
        optimizer.step() #aktualisiert Gewichte
    #Training beendet
        
    model.eval()
    y_pred = model(test_input)
    after_train = criterion(y_pred.squeeze(), test_output_tensor)
    #print('Test loss after Training' , after_train.item())
    
    #Ergebnisse plotten
    teams= list(test_output.iloc[:,:1].values)
    errors = np.array(errors)
    plt.figure(figsize=(12, 5))
    graf01 = plt.subplot(1, 2, 1) # nrows, ncols, index
    graf01.set_title('Errors')
    plt.plot(errors, '-')
    graf02 = plt.subplot(1, 2, 2)
    graf02.set_title('Tests')
    a = plt.plot(test_output_tensor.numpy()[0], 'yo', label='Real')
    plt.setp(a, markersize=10)
    for i in range(1,18):
        a= plt.plot(i,test_output_tensor.numpy()[i],'yo')
        #a=plt.annotate(teams[i][0],(i*20,test_output_tensor.numpy()[i]))
        plt.setp(a, markersize=10)
    
    a = plt.plot(y_pred.detach().numpy()[0], 'b+', label='Predicted')
    plt.setp(a, markersize=10)
    for i in range(1,18):
        a = plt.plot(i,y_pred.detach().numpy()[i], 'b+')
        plt.setp(a, markersize=10)
    plt.legend(loc=7)
    plt.show()
    
def test_training(dfmod):
    training= dfmod.iloc[:-18, :-1]
    test_withteam= dfmod.iloc[-18:]
    test=dfmod.iloc[-18:,:-1]
    #print(dfmod)
    #print(dfmod.iloc[:-18])
    #Alles Normalisieren zwischen 0 und 1
    #print(test)
    #print(training)
    training,test= normalize(training,test)
    test_withteam.iloc[-18:,:-1]=test
    
    #Trainngs und Testssets mische 
    training= training.sample(frac=1)
    test_withteam= test_withteam.sample(frac=1)
    
    #Aufteilung in Eingabe und Ausgabe
    training_input= training.iloc[:,:-3]
    training_output= training.iloc[:,-3:]
    test_input= test_withteam.iloc[:,:-4]
    test_output= test_withteam.iloc[:,-4:]

    #Win, Draw, Defeat umkonvertieren in Sieg ode nicht Sieg, oder Draw oder nicht Draw..
    #Von 3 Klassen zu zwei konvertieren
    
    training_output= win(training_output)
    test_values=win(test_output.iloc[:,:-1])
    test_output= test_output.iloc[:,3:]
    test_output["win"]=test_values
    #print(test_output)
 
    return training_input,training_output,test_input, test_output
        

    
def win(data):
    copy= data.copy()
    copy['new']= 2
    for i, rows in copy.iterrows():
        if rows['Win'] == 1:
            rows['new'] = 1
        if rows['Draw'] == 1:
            rows['new'] = 0
        if rows['Defeat'] == 1:
            rows['new'] = 0
    return copy.iloc[:, -1] 

def draw(data):
    copy= data.copy()
    copy['new']= 2
    for i, rows in copy.iterrows():
        if rows['Win'] == 1:
            rows['new'] = 0
        if rows['Draw'] == 1:
            rows['new'] = 1
        if rows['Defeat'] == 1:
            rows['new'] = 0
    return copy.iloc[:, -1] 

def defeat(data):
    copy= data.copy()
    copy['new']= 2 
    for i, rows in copy.iterrows():
        if rows['Win'] == 1:
            rows['new'] = 0
        if rows['Draw'] == 1:
            rows['new'] = 0
        if rows['Defeat'] == 1:
            rows['new'] = 1
    return copy.iloc[:, -1] 

def normalize(training,test):
    #normalize the data between 0 and 1
    column_names=list(training)[:-3]   
    #max finden und dividieren
    count=0
    for i in column_names:
        column_trainng= training[i]
        column_test= test[i]
        num= max(column_trainng.max(), column_test.max())
    
        if int(num)<10:
            training.iloc[:,count] /=10
            test.iloc[:,count]/=10
        elif int(num)<100:
            training.iloc[:,count]/=100
            test.iloc[:,count]/=100
        elif int(num)<1000:
            training.iloc[:,count]/=1000
            test.iloc[:,count]/=1000
        count= count+1
    return training,test


def sammeln(spieldaten,anfang,ende,feature):
    anfang=anfang*9
    ende= ende*9+9
    dfmod= ""
    count=0
    mannschaften=[]
    tore=[]
    torediff=[]
    schuesse=[]
    laufen=[]
    paesse=[]
    erfolg_paesse=[]
    fehlpaesse=[]
    passquote=[]
    ballbesitz=[]
    zweikampfquote=[]
    foul=[]
    gefoult=[]
    abseits=[]
    ecken=[]
    abseits=[]
    eff=[]
    win=[]
    draw=[]
    defeat=[]
    with (open(spieldaten, "rb")) as openfile:
        while True:
            try:
                if count< ende:
                    spieltag= pickle.load(openfile)
                    #print(spieltag)
                    #print(spieltag[1], spieltag[2], spieltag[4][0:2], spieltag[5][0:2])
                    if spieltag[1]=="Mannschaft1":
                        continue
                    if count >= anfang:
                        #print(spieltag)
                        mannschaften.append(spieltag[1])
                        mannschaften.append(spieltag[2])
                        torediff.append(int(spieltag[4][0]))
                        torediff.append(int(spieltag[5][0]))
                        for i in feature:
                            if i=="effizienz":
                                eff.append(int(spieltag[4][0])/int(spieltag[4][1]))
                                eff.append(int(spieltag[5][0])/int(spieltag[5][1]))
                            if i== "tore":    
                                tore.append(int(spieltag[4][0]))
                                tore.append(int(spieltag[5][0]))

                            elif i == "schuesse":
                                schuesse.append(int(spieltag[4][1]))
                                schuesse.append(int(spieltag[5][1]))
                            elif i== "laufen":
                                laufen.append(float((spieltag[4][2][:-2]).replace(",",".")))
                                laufen.append(float((spieltag[5][2][:-2]).replace(",",".")))
                            elif i== "paesse":
                                paesse.append(int(spieltag[4][3]))
                                paesse.append(int(spieltag[5][3]))
                            elif i== "epaesse":
                                erfolg_paesse.append(int(spieltag[4][4]))
                                erfolg_paesse.append(int(spieltag[5][4]))
                            elif i== "fehlpaesse":
                                fehlpaesse.append(int(spieltag[4][5]))
                                fehlpaesse.append(int(spieltag[5][5]))
                            elif i== "passquote":
                                passquote.append(int(spieltag[4][6][:-1]))
                                passquote.append(int(spieltag[5][6][:-1]))
                            elif i== "ballbesitz":
                                ballbesitz.append(int(spieltag[4][7][:-1]))
                                ballbesitz.append(int(spieltag[5][7][:-1]))
                            elif i=="zweikampf":
                                zweikampfquote.append(int(spieltag[4][8][:-1]))
                                zweikampfquote.append(int(spieltag[5][8][:-1]))
                            elif i=="foul":
                                foul.append(int(spieltag[4][9]))
                                foul.append(int(spieltag[5][9]))
                            elif i=="gfoul":
                                gefoult.append(int(spieltag[4][10]))
                                gefoult.append(int(spieltag[5][10]))
                            elif i=="abseits":
                                abseits.append(int(spieltag[4][11]))
                                abseits.append(int(spieltag[5][11]))
                            elif i=="ecken":
                                ecken.append(int(spieltag[4][12]))
                                ecken.append(int(spieltag[5][12]))
                        #letzten Drei Spalten    
                        differenz= torediff[-2:]
                        diff=int(differenz[0])-int(differenz[1])
                        if diff >0:
                            win.append(1)
                            win.append(0)
                            draw.append(0)
                            draw.append(0)
                            defeat.append(0)
                            defeat.append(1)
                        elif diff==0:
                            win.append(0)
                            win.append(0)
                            draw.append(1)
                            draw.append(1)
                            defeat.append(0)
                            defeat.append(0)
                        elif diff <0:
                            win.append(0)
                            win.append(1)
                            draw.append(0)
                            draw.append(0)
                            defeat.append(1)
                            defeat.append(0)
                        data={}
                        for i in feature:
                            if i=="effizienz":
                                data.update({i:eff})
                            if i=="tore":
                                data.update({i:tore})
                            if i=="schuesse":
                                data.update({i:schuesse})
                            if i=="laufen":
                                data.update({i:laufen})
                            if i=="paesse":
                                data.update({i:paesse})
                            if i=="epaesse":
                                data.update({i:erfolg_paesse})
                            if i=="fehlpaesse":
                                data.update({i:fehlpaesse})
                            if i=="passquote":
                                data.update({i:passquote})
                            if i=="ballbesitz":
                                data.update({i:ballbesitz})
                            if i=="zweikampf":
                                data.update({i:zweikampfquote})
                            if i=="foul":
                                data.update({i:foul})
                            if i=="gfoul":
                                data.update({i:gefoult})
                            if i=="abseits":
                                data.update({i:abseits})
                            if i=="ecken":
                                data.update({i:ecken})
                        data.update({"Win":win})
                        data.update({"Draw":draw})
                        data.update({"Defeat":defeat})
                        data.update({"Mannschaften":mannschaften})
                     

                        dfmod= pd.DataFrame(data)
                    
                        
                    count=count+1
                else:
                    break
            except EOFError:
                break
    
    return dfmod

#Feedfoward-Nertzwerkmodell erstellen
class Net(torch.nn.Module):
    def __init__(self, input_size, hidden_size):
        super(Net, self).__init__()
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.fc1 = torch.nn.Linear(self.input_size, self.hidden_size)
        self.relu = torch.nn.ReLU()
        self.fc2 = torch.nn.Linear(self.hidden_size, 1)
        self.sigmoid = torch.nn.Sigmoid() 
    def forward(self, x):
        hidden = self.fc1(x)
        relu = self.relu(hidden)
        output = self.fc2(relu)
        output = self.sigmoid(output)
        return output

def vorsaison(daten,anfang,ende, feature):
    if anfang <0:    
        vanfang=34+anfang
        anfang=0
        vende=34
        if int(daten[-5])==0:
            vspieldaten= daten[:-6]+str((int(daten[-6])-1))+str(9)+".pkl"
        else:
            vspieldaten= daten[:-5]+str(int(daten[-5])-1)+".pkl"
        vmannschaften=einlesen_mannschaften(vspieldaten)
        mannschaften= einlesen_mannschaften(daten)
        aufsteiger=set(mannschaften)-set(vmannschaften)
        vaufsteiger= set(vmannschaften)-set(mannschaften)
        
        spiele=sammeln(daten,anfang,ende,feature)
        for i in aufsteiger:
            spiele= spiele[spiele.Mannschaften!=i]

        vspiele= sammeln(vspieldaten,vanfang,vende,feature)
        for i in vaufsteiger:
            vspiele= vspiele[vspiele.Mannschaften!=i]

        dfmodhelp= einlesen_nextgegner(daten,ende,ende+1)
        aufsteiger= list(aufsteiger)
        nextgegner=[]
        for i in dfmodhelp:
            for j in aufsteiger:
                if j in i:
                    i.remove(j)
                    nextgegner.append(i[0])
        allespiele=pd.concat([vspiele,spiele])
        for i in nextgegner:
            allespiele=allespiele[allespiele.Mannschaften!=i]
        
        training_input,training_output,test_input,test_output= test_training(allespiele)
    
        # in Tensoren umwandeln    
        training_input = torch.FloatTensor(training_input.values) 
        training_output = torch.FloatTensor(training_output.values) 
        test_input = torch.FloatTensor(test_input.values) 
        test_output_tensor = torch.FloatTensor(test_output.iloc[:,1:].values)

        #Eingabedimension= Anzahl der Features
        input_size= training_input.size()[1]
        hidden_size= input_size+2 #Anzahl der Knoten
        model= Net(input_size,hidden_size)#Modell erstellen
        criterion = torch.nn.BCELoss() #Lossfunktion: BCE da wir bin채re Ergebnisse haben

        #optimizer: enthaelt aktuellen Status und aktualisiert die Parameter basierend 
        #auf den berechneten Verl채ufen
        #ohne momentum paramter
        optimizer = torch.optim.SGD(model.parameters(), lr = 0.9) 
        #mit momentum parameter
        #optimizer = torch.optim.SGD(model.parameters(), lr = 0.9, momentum=0.2)

        #Trainingsanfang

        model.eval()#Testmodus, keine Gewichte werden gelernt, nur Verlust angezeigt
        y_pred = model(test_input)
        before_train = criterion(y_pred.squeeze(), test_output_tensor)
        #print('Test loss before training' , before_train.item())

        model.train()
        epochen = 5000
        errors = []
        for epoch in range(epochen):
            optimizer.zero_grad()
            # Forward pass
            y_pred = model(training_input)
            # Verlustberechnung
            loss = criterion(y_pred.squeeze(), training_output)
            errors.append(loss.item())
            #print('Epoch {}: train loss: {}'.format(epoch, loss.item()))

            # Backward berechnet die Gradienten
            loss.backward()
            optimizer.step() #aktualisiert Gewichte
        #Training beendet

        model.eval()
        y_pred = model(test_input)
        after_train = criterion(y_pred.squeeze(), test_output_tensor)
        #print('Test loss after Training' , after_train.item())

        #Ergebnisse plotten
        teams= list(test_output.iloc[:,:1].values)
        errors = np.array(errors)
        plt.figure(figsize=(12, 5))
        graf01 = plt.subplot(1, 2, 1) # nrows, ncols, index
        graf01.set_title('Errors')
        plt.plot(errors, '-')
        graf02 = plt.subplot(1, 2, 2)
        graf02.set_title('Tests')
        a = plt.plot(test_output_tensor.numpy()[0], 'yo', label='Real')
        plt.setp(a, markersize=10)
        for i in range(1,18):
            a= plt.plot(i,test_output_tensor.numpy()[i],'yo')
            #a=plt.annotate(teams[i][0],(i*20,test_output_tensor.numpy()[i]))
            plt.setp(a, markersize=10)

        a = plt.plot(y_pred.detach().numpy()[0], 'b+', label='Predicted')
        plt.setp(a, markersize=10)
        for i in range(1,18):
            a = plt.plot(i,y_pred.detach().numpy()[i], 'b+')
            plt.setp(a, markersize=10)
        plt.legend(loc=7)
        plt.show()
            

def einlesen_nextgegner(spieldaten,anfang,ende):

    anfang=anfang*9
    ende= ende*9
    
    dfmod= ""
    count=0
    games=[]
    with (open(spieldaten, "rb")) as openfile:
        while True:
            try:
                if count< ende:
                    spieltag= pickle.load(openfile)
                    #print(spieltag)
                    #print(spieltag[1], spieltag[2], spieltag[4][0:2], spieltag[5][0:2])
                    if spieltag[1]=="Mannschaft1":
                        continue
                    if count >= anfang:
                        
                        mannschaften=[]
                        mannschaften.append(spieltag[1])
                        mannschaften.append(spieltag[2])
                        games.append(mannschaften)
                        
                else:
                    break
                count=count+1
            except EOFError:
                break
    return games
                    
#vorsaison("Statistik2019.pkl",-1,6,i)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier

def einlesen(spieldaten,anfang,ende,anfrage):    
    mannschaft1=[]
    mannschaft2=[]
    tore= []
    schuesse=[]
    anfrageliste_e=[]
    anfrageliste_f=[]
    anfrageliste_b=[]
    anfrageliste_p=[]
    anfrageliste_l=[]
    anfrageliste_z=[]
    anfrageliste_s=[]
    anfrageliste_paesse=[]
    anfrageliste_ep=[]
    anfrageliste_fp=[]
    anfrageliste_gfoul=[]
    anfrageliste_a=[]
    anfrageliste_ecken=[]
    anfrageliste_komplett=[]
    anfrageliste_tore=[]
    ergebnisse=[]

    count=0
    dfmod= ""
    with (open(spieldaten, "rb")) as openfile:
        while True:
            try:
                if count< ende:
                    spieltag= pickle.load(openfile)
                    #print(spieltag)
                    #print(spieltag[1], spieltag[2], spieltag[4][0:2], spieltag[5][0:2])
                    if spieltag[1]=="Mannschaft1":
                        continue
                    if count >= anfang:
                        mannschaft1.append(spieltag[1])
                        mannschaft2.append(spieltag[2])
                        for i in anfrage:
                            #print(i)
                            if i=="effizienz":
                                eff_heim= int(spieltag[4][0])/int(spieltag[4][1])
                                eff_gast= int(spieltag[5][0])/int(spieltag[5][1])
                                effi_gesamt= eff_heim-eff_gast
                                #print(effi_gesamt)
                                anfrageliste_e.append(effi_gesamt)
                            elif i=="ballbesitz":
                                ballbesitz_heim= int(spieltag[4][7][:-1])
                                ballbesitz_gast= int(spieltag[5][7][:-1])
                                ballbesitz_gesamt= ballbesitz_heim/ballbesitz_gast
                                anfrageliste_b.append(ballbesitz_gesamt)
                            elif i=="tore":
                                tore_heim= int(spieltag[4][0])
                                tore_gast= int(spieltag[5][0])
                                tore_gesamt= tore_heim-tore_gast
                                anfrageliste_tore.append(tore_gesamt)
                                
                            elif i=="schuesse":
                                schuesse_heim= int(spieltag[4][1])
                                schuesse_gast= int(spieltag[5][1])
                                schuesse_gesamt= schuesse_heim/schuesse_gast
                                anfrageliste_s.append(schuesse_gesamt)
                            elif i=="paesse":
                                paesse_heim= int(spieltag[4][3])
                                paesse_gast= int(spieltag[5][3])
                                paesse_gesamt= paesse_heim/paesse_gast
                                anfrageliste_paesse.append(paesse_gesamt)
                            elif i=="epaesse":
                                epaesse_heim= int(spieltag[4][4])
                                epaesse_gast= int(spieltag[5][4])
                                epaesse_gesamt= epaesse_heim/epaesse_gast
                                anfrageliste_ep.append(epaesse_gesamt)
                            elif i=="fpaesse":
                                fpaesse_heim= int(spieltag[4][5])
                                fpaesse_gast= int(spieltag[5][5])
                                fpaesse_gesamt= fpaesse_heim/fpaesse_gast
                                anfrageliste_fp.append(fpaesse_gesamt)
                            elif i=="foul":
                                foul_heim= int(spieltag[4][9])
                                foul_gast= int(spieltag[5][9])
                                foul_gesamt= foul_heim/foul_gast
                                anfrageliste_f.append(foul_gesamt)
                            elif i=="gfoul":
                                gfoul_heim= int(spieltag[4][10])
                                gfoul_gast= int(spieltag[5][10])
                                gfoul_gesamt= gfoul_heim/gfoul_gast
                                anfrageliste_gfoul.append(gfoul_gesamt)
                            elif i=="abseits":
                                a_heim= int(spieltag[4][11])
                                a_gast= int(spieltag[5][11])
                                if a_gast==0:
                                    a_gast=1
                                if a_heim==0:
                                    a_heim=1
                                
                                a_gesamt= a_heim/a_gast
                                anfrageliste_a.append(a_gesamt)
                            elif i=="ecken":
                                ecken_heim= int(spieltag[4][12])
                                ecken_gast= int(spieltag[5][12])
                                if ecken_heim==0:
                                    ecken_heim=1
                                if ecken_gast==0:
                                    ecken_gast=1
                                ecken_gesamt= ecken_heim/ecken_gast
                                anfrageliste_ecken.append(ecken_gesamt)
                            elif i=="laufen":
                                km_heim= float((spieltag[4][2][:-2]).replace(",","."))
                                km_gast= float((spieltag[5][2][:-2]).replace(",","."))
                                lauf_gesamt= km_heim/km_gast
                                anfrageliste_l.append(lauf_gesamt)
                            elif i=="passen":
                                passen_heim= int(spieltag[4][6][:-1])
                                passen_gast= int(spieltag[5][6][:-1])
                                passen_gesamt= passen_heim/passen_gast
                                anfrageliste_p.append(passen_gesamt)
                                #print(passen_gesamt)
                            elif i=="zweikampf":
                                zweikampf_heim= int(spieltag[4][8][:-1])
                                zweikampf_gast= int(spieltag[5][8][:-1])
                                zweikampf_gesamt= zweikampf_heim/zweikampf_gast
                                anfrageliste_z.append(zweikampf_gesamt)
                            
            
                        ergebnis= (spieltag[4][0:1][0])+":"+(spieltag[5][0:1][0])
                        ergebnisse.append(ergebnis)
                        data= {"Mannschaft1":mannschaft1,"Mannschaft2":mannschaft2}
                        #print(anfrageliste_komplett)
                        for i in anfrage:
                            if i=="effizienz":
                                data.update({i:anfrageliste_e})
                            if i=="foul":
                                data.update({i:anfrageliste_f})
                            if i=="tore":
                                data.update({i:anfrageliste_tore})
                            if i=="laufen":
                                data.update({i:anfrageliste_l})
                            if i=="passen":
                                data.update({i:anfrageliste_p})
                            if i=="zweikampf":
                                data.update({i:anfrageliste_z})
                            if i=="schuesse":
                                data.update({i:anfrageliste_s})
                            if i=="paesse":
                                data.update({i:anfrageliste_paesse})
                            if i=="epaesse":
                                data.update({i:anfrageliste_ep})
                            if i=="fpaesse":
                                data.update({i:anfrageliste_fp})
                            if i=="ballbesitz":
                                data.update({i:anfrageliste_b})
                            if i=="gfoul":
                                data.update({i:anfrageliste_gfoul})
                            if i=="abseits":
                                data.update({i:anfrageliste_a})
                            if i=="ecken":
                                data.update({i:anfrageliste_ecken})
                            
                        #print(data)

                        dfmod= pd.DataFrame(data)
                        
                    count=count+1
                else:
                    break
            except EOFError:
                break
    return dfmod, mannschaft1,mannschaft2,ergebnisse

tanfang=[10,20,30,8,18,28,6,16,26,1,11,21]
tende=[11,21,31,11,21,31,11,21,31,11,21,31]

def RandomForest(spieldaten, anfang,ende,anfrage):
    
    anfang=anfang*9
    ende=ende*9 
    dfmod,m1,m2,e1= einlesen(spieldaten,anfang,ende,anfrage)
    X=dfmod.drop(["Mannschaft1","Mannschaft2"],axis=1)
    y=[]
    for i in e1:
        numbers= str(i).split(":")
        diff = int(numbers[0])-int(numbers[1])
        if diff>0:
            y.append(1)
        elif diff==0:
            y.append(0)
        elif diff<0:
            y.append(-1)
    y= np.array(y)
    model = RandomForestClassifier(n_estimators=5000, max_features=len(anfrage))
    # fit the model on the whole dataset
    model.fit(X, y)
    #print(model.feature_importances_)
    return model.feature_importances_

tanfrage=["tore","effizienz","schuesse","ecken","ballbesitz","gfoul","paesse","epaesse","abseits","fpaesse","foul","passen","zweikampf","laufen"]
for i in range(0,len(tanfang)):
    
    anfang_neu=tanfang[i]
    ende_neu=tende[i]
    liste= RandomForest("Statistik2019.pkl", anfang_neu,ende_neu,tanfrage)
    features={}
    for i in range(0,len(liste)):
        features[tanfrage[i]]=liste[i]
    sortierte_features=(sorted(features.items(), key=lambda x: x[1],reverse=True))
    tanfrage_neu=[]
    for i in sortierte_features:
        tanfrage_neu.append(i[0])
        
    tanfrage_neu.pop()
    tanfrage_neu.pop()
    tanfrage_neu.pop()
    print("Spiele von "+str(anfang_neu)+" bis "+str(ende_neu)+"\n")
    main("Statistik2019.pkl",anfang_neu,ende_neu,tanfrage_neu)
    
anfrageliste=[["passquote","paesse","zweikampf","epaesse","ballbesitz","schuesse","tore","effizienz","laufen","fehlpaesse","foul","gfoul"]]
feature_1=["fehlpaesse","foul","gfoul","abseits","ecken","tore"]
'''print("#################### 1 FEATURE ##################")
for j in feature_1:
    i=[]
    i.append(j)
    print("ANZAHL DER SPIELE 1"+"\n")
    main("Statistik2019.pkl",10,11,i)
    main("Statistik2019.pkl",20,21,i)
    main("Statistik2019.pkl",30,31,i)
    print("ANZAHL DER SPIELE 3"+"/n")
    main("Statistik2019.pkl",8,11,i)
    main("Statistik2019.pkl",18,21,i)
    main("Statistik2019.pkl",28,31,i)
    print("ANZAHL DER SPIELE 5"+"/n")
    main("Statistik2019.pkl",6,11,i)
    main("Statistik2019.pkl",16,21,i)
    main("Statistik2019.pkl",26,31,i)
    print("ANZAHL DER SPIELE 10"+"/n")
    main("Statistik2019.pkl",1,11,i)
    main("Statistik2019.pkl",11,21,i)
    main("Statistik2019.pkl",21,31,i)'''
```
