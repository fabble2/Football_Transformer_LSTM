```python
import pandas as pd
import pickle
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
def main():
    dfmod=sammeln("Statistik2019.pkl",18,22)
    training= dfmod.iloc[:-18]
    test= dfmod.iloc[-18:]
    #print(test)
    
    #normalize the data between 0 and 1
    column_names=list(training)[:-3]   
    #max finden und dividieren
    count=0
    for i in column_names:
        column_trainng= training[i]
        column_test= test[i]
        num= max(column_trainng.max(), column_test.max())
    
        if int(num)<10:
            training.iloc[:,count] /=10
            test.iloc[:,count]/=10
        elif int(num)<100:
            training.iloc[:,count]/=100
            test.iloc[:,count]/=100
        elif int(num)<1000:
            training.iloc[:,count]/=1000
            test.iloc[:,count]/=1000
        count= count+1

    #Trainngs und Testssets mische 
    training= training.sample(frac=1)
    test= test.sample(frac=1)
    #Aufteilung in Eingabe und Ausgabe
    training_input= training.iloc[:,:-3]
    training_output= training.iloc[:,-3:]
    test_input= test.iloc[:,:-3]
    test_output= test.iloc[:,-3:]
    #print(test_output)
    training_output= convert_win(training_output)
    test_output=convert_win(test_output)
    
    # in Tensoren umwandeln 
    training_input = torch.FloatTensor(training_input.values) 
    training_output = torch.FloatTensor(training_output.values) 
    test_input = torch.FloatTensor(test_input.values) 
    test_output = torch.FloatTensor(test_output.values)
    
    input_size= training_input.size()[1]
    hidden_size= 16
    model= Net(input_size,hidden_size)
    criterion = torch.nn.BCELoss() # works for binary classification
    # without momentum parameter
    optimizer = torch.optim.SGD(model.parameters(), lr = 0.9) 
    #with momentum parameter
    optimizer = torch.optim.SGD(model.parameters(), lr = 0.9, momentum=0.2)
    
    model.eval()
    y_pred = model(test_input)
    before_train = criterion(y_pred.squeeze(), test_output)
    print('Test loss before training' , before_train.item())
    
    model.train()
    epochs = 5000
    errors = []
    for epoch in range(epochs):
        optimizer.zero_grad()
        # Forward pass
        y_pred = model(training_input)
        # Compute Loss
        loss = criterion(y_pred.squeeze(), training_output)
        errors.append(loss.item())
        #print('Epoch {}: train loss: {}'.format(epoch, loss.item()))
        # Backward pass
        loss.backward()
        optimizer.step()
    model.eval()
    y_pred = model(test_input)
    after_train = criterion(y_pred.squeeze(), test_output)
    #print('Test loss after Training' , after_train.item())
    
    errors = np.array(errors)
    plt.figure(figsize=(12, 5))
    graf02 = plt.subplot(1, 2, 1) # nrows, ncols, index
    graf02.set_title('Errors')
    plt.plot(errors, '-')
    plt.xlabel('Epochs')
    graf03 = plt.subplot(1, 2, 2)
    graf03.set_title('Tests')
    a = plt.plot(test_output.numpy(), 'yo', label='Real')
    plt.setp(a, markersize=10)
    a = plt.plot(y_pred.detach().numpy(), 'b+', label='Predicted')
    plt.setp(a, markersize=10)
    plt.legend(loc=7)
    plt.show()
        

    
def convert_win(data):
    copy= data.copy()
    copy['new']= 2 #create a new column with any value
    for i, rows in copy.iterrows():
        if rows['Win'] == 1:
            rows['new'] = 1
        if rows['Draw'] == 1:
            rows['new'] = 0
        if rows['Defeat'] == 1:
            rows['new'] = 0
    return copy.iloc[:, -1] 

main()
```

```python
def sammeln(spieldaten,anfang,ende):
    anfang=anfang*9
    ende= ende*9
    dfmod= ""
    count=0
    tore=[]
    schuesse=[]
    laufen=[]
    paesse=[]
    erfolg_paesse=[]
    fehlpaesse=[]
    passquote=[]
    ballbesitz=[]
    zweikampfquote=[]
    foul=[]
    gefoult=[]
    abseits=[]
    ecken=[]
    abseits=[]
    win=[]
    draw=[]
    defeat=[]
    with (open(spieldaten, "rb")) as openfile:
        while True:
            try:
                if count< ende:
                    spieltag= pickle.load(openfile)
                    #print(spieltag)
                    #print(spieltag[1], spieltag[2], spieltag[4][0:2], spieltag[5][0:2])
                    if spieltag[1]=="Mannschaft1":
                        continue
                    if count >= anfang:
                        #print(spieltag[3])
                        for i in range(4,6):
                            tore.append(int(spieltag[i][0]))
                            schuesse.append(int(spieltag[i][1]))
                            laufen.append(float((spieltag[i][2][:-2]).replace(",",".")))
                            paesse.append(int(spieltag[i][3]))
                            erfolg_paesse.append(int(spieltag[i][4]))
                            fehlpaesse.append(int(spieltag[i][5]))
                            passquote.append(int(spieltag[i][6][:-1]))
                            ballbesitz.append(int(spieltag[i][7][:-1]))
                            zweikampfquote.append(int(spieltag[i][8][:-1]))
                            foul.append(int(spieltag[i][9]))
                            gefoult.append(int(spieltag[i][10]))
                            abseits.append(int(spieltag[i][11]))
                            ecken.append(int(spieltag[i][12]))
                        
                        #letzten Drei Spalten    
                        differenz= tore[-2:]
                        diff=int(differenz[0])-int(differenz[1])
                        if diff >0:
                            win.append(1)
                            win.append(0)
                            draw.append(0)
                            draw.append(0)
                            defeat.append(0)
                            defeat.append(1)
                        elif diff==0:
                            win.append(0)
                            win.append(0)
                            draw.append(1)
                            draw.append(1)
                            defeat.append(0)
                            defeat.append(0)
                        elif diff <0:
                            win.append(0)
                            win.append(1)
                            draw.append(0)
                            draw.append(0)
                            defeat.append(1)
                            defeat.append(0)
                            
                        data= {"Tore":tore,"Schuesse":schuesse,"Laufen":laufen, "Paesse":paesse,"erfolgPaesse":erfolg_paesse,
      "Fehlpaesse":fehlpaesse,"Passquote":passquote,"Ballbesitz":ballbesitz,"Zweikampf":zweikampfquote,
      "Foul":foul,"Gefoult":gefoult,"Abseits":abseits,"Ecken":ecken,"Win":win,"Draw":draw,"Defeat":defeat}
                        #print(data)

                        dfmod= pd.DataFrame(data)
                    
                        
                    count=count+1
                else:
                    break
            except EOFError:
                break
    return dfmod
    import torch
class Net(torch.nn.Module):
    def __init__(self, input_size, hidden_size):
        super(Net, self).__init__()
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.fc1 = torch.nn.Linear(self.input_size, self.hidden_size)
        self.relu = torch.nn.ReLU()
        self.fc2 = torch.nn.Linear(self.hidden_size, 1)
        self.sigmoid = torch.nn.Sigmoid() 
    def forward(self, x):
        hidden = self.fc1(x)
        relu = self.relu(hidden)
        output = self.fc2(relu)
        output = self.sigmoid(output)
        return output
```

```python
import torch
class Net(torch.nn.Module):
    def __init__(self, input_size, hidden_size):
        super(Net, self).__init__()
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.fc1 = torch.nn.Linear(self.input_size, self.hidden_size)
        self.relu = torch.nn.ReLU()
        self.fc2 = torch.nn.Linear(self.hidden_size, 1)
        self.sigmoid = torch.nn.Sigmoid() 
    def forward(self, x):
        hidden = self.fc1(x)
        relu = self.relu(hidden)
        output = self.fc2(relu)
        output = self.sigmoid(output)
        return output
```
